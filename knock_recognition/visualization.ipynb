```python
import os
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
from pydub import AudioSegment
from pydub.playback import play
import time

# For video writer.
import cv2

# For progress bar.
import tqdm

# For wav file. 
import io
import scipy.io.wavfile

# Spectrogram.
import scipy.signal
import librosa
import librosa.display
```

```python
# Load audio.
input_audio = "./data/raw/demo_audio.mp3"
sound = AudioSegment.from_mp3(input_audio)
sound = sound.set_channels(1)
sample_rate = sound.frame_rate
audio_samples = sound.get_array_of_samples()
audio_data_mono = np.array(audio_samples).T.astype(np.float32)
audio_data_mono /= np.iinfo(audio_samples.typecode).max

# Slicing.
idx_start_sec = 84.0
idx_end_sec = 97.0
idx_start = int(idx_start_sec * sample_rate)
idx_end = int(idx_end_sec * sample_rate)
audio_data_mono = audio_data_mono[idx_start:idx_end]

# Print info.
total_samples = len(audio_data_mono)
total_sec = total_samples / sample_rate
print("Sample rate: ", sample_rate)
print("Total samples: ", total_samples)
print("Duration: {} sec".format(total_sec))
```

```python
# Show overall. 
fig, ax = plt.subplots(figsize=[20, 4], dpi=300)

# Frame off.
fig.patch.set_visible(False)
# ax.set_xticks([])
# ax.set_yticks([])
# for spine in ax.spines:
#     ax.spines[spine].set_visible(False)

time_ticks = np.arange(0, total_samples / sample_rate + 1, 1).astype(int)
x_ticks = time_ticks * sample_rate
ax.set_xticks(x_ticks)
ax.set_xticklabels(time_ticks, fontsize=20)
ax.set_yticks([-0.15, 0, 0.15])
ax.set_yticklabels([-0.15, 0, 0.15], fontsize=20)
ax.set_xlim([0, x_ticks.max()])
ax.set_ylim([-0.15, 0.15])
ax.set_xlabel("Time (sec)", fontsize=20)
ax.set_ylabel("Amplitude", fontsize=20)

# Plot.
ax.plot(audio_data_mono, color="#6679e3")
```

```python
# Spectrogram.
ps = librosa.feature.melspectrogram(y=audio_data_mono, sr=sample_rate)
ps_db = librosa.power_to_db(ps, ref=np.max)
fig, ax = plt.subplots(figsize=[20,4], dpi=300)
# ax.imshow(ps_db, origin="lower", cmap=plt.get_cmap("magma"))

# Frame off.
# fig.patch.set_visible(False)
# ax.set_xticks([])
# ax.set_yticks([])
# for spine in ax.spines:
#     ax.spines[spine].set_visible(False)
librosa.display.specshow(ps_db, x_axis='s', y_axis="mel", auto_aspect=False, ax=ax)

# Set axis.
ax.set_xticks([])
ax.set_xlabel("")
ax.set_ylabel("Frequency (Hz)", fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
```
